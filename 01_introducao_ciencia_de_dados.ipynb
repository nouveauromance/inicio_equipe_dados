<h1 style="color: #f686bd">O que é ciência de dados?</h1>
<div style="display: flex; align-items: start; gap: 20px;">
<img src="imagens/DataScience.png" width="400">
<div>
A ciência de dados é uma área multidisciplinar que utiliza métodos científicos. Ela mistura matemática e estatística, programação especializada, análise de dados avançada, inteligência artificial (IA) e aprendizado de máquina, com conhecimento sobre negócios para encontrar respostas valiosas que estão escondidas no meio de um mar de informações. No fim, as empresas usam essas descobertas para fazer escolhas mais inteligentes e planejar o futuro com segurança.
<br/> <br/> <br/>
<h4 style="color: #f686bd">Por que a Ciência de Dados está tão em alta?</h4>
Hoje em dia, tudo o que fazemos gera uma montanha de informações (dados). Como o volume de dados não para de crescer, a Ciência de Dados se tornou uma das áreas que mais evolui no mundo. As empresas dependem de profissionais para traduzir toda essa bagunça de informações em conselhos práticos do que elas devem fazer para ter mais sucesso e lucrar mais.

<h2 style="color: #f686bd">O Ciclo de Vida da Ciência de Dados (Passo a Passo)</h2>

- <span style="color: #f686bd">1. Ingestão de Dados: </span>Tudo começa reunindo as informações. Os dados vêm de todos os lugares e de todas as formas. É o ato de buscar e capturar dados de diversas fontes. Podem ser planilhas organizadas, textos digitados à mão, áudios, imagens de exames ou até informações em tempo real que vêm de um relógio inteligente (IoT).

- <span style="color: #f686bd">2. Armazenamento e Processamento: </span>É a fase de organizar e "limpar" os dados. Aqui entra o famoso [ETL](https://www.ibm.com/br-pt/think/topics/etl) (Extrair, Transformar e Carregar). É tirar informações duplicadas, corrigir erros (como alguém que digitou "300 anos" na idade) e guardar tudo em um local seguro e organizado (como um [Data Warehouse ou Data Lake](https://www.ibm.com/br-pt/think/topics/data-warehouse-vs-data-lake-vs-data-lakehouse)).

- <span style="color: #f686bd">3. Engenharia de Atributos / Feature Engineering: </span>Às vezes, os dados que temos não são suficientes por si só. Precisamos combinar informações para criar "pistas" melhores para o computador entender o problema. É usar a criatividade e a matemática para criar novas colunas de dados a partir das colunas que já existem.

- <span style="color: #f686bd">4. Análise de Dados: </span>Primeiro, fazemos uma "análise exploratória" para entender o que temos nas mãos (ex: qual a média de idade? Quantos homens e mulheres?).

- <span style="color: #f686bd">5. Modelagem / Machine Learning: </span> É aqui que a Inteligência Artificial entra em cena de verdade. É a etapa de escolher o "motor" matemático que vai resolver o problema. É a aplicação de algoritmos que vão ler os dados de Treino e criar regras matemáticas automáticas. Existem vários tipos de algoritmos para problemas diferentes.

- <span style="color: #f686bd">6. Comunicação: </span> Nenhum diretor de empresa quer ler linhas de código de programação ou fórmulas de estatística. Eles precisam de respostas claras. É a tradução de todo aquele trabalho técnico em gráficos, painéis de controle (dashboards) e relatórios visuais fáceis de entender. Para isso, usamos linguagens como Python ou ferramentas visuais específicas.
