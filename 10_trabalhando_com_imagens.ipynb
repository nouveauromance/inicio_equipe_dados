### 🌈 O que são canais de cor?
As letras representam:

- R (Red) → vermelho
- G (Green) → verde
- B (Blue) → azul

Cada pixel da imagem é formado por esses 3 valores. [255, 0, 0] = muito vermelho, zero verde e zero azul. => com essas misturas criamos as cores.

Voltando para o problema das bibliotecas *BGR vs RGB*
- cv2 (OpenCV) usa: [Blue, Green, Red]  # BGR
- matplotlib (plt) espera: [Red, Green, Blue]  # RGB

para resolver vamos fazer uma conversão:



```python
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.axis('off') #tira os numeros do lado da imagem
```

> Porque de 0 a 255 ? As matrizes numpy são tipo `uint8` Isso significa **Unsigned Integer de 8 bits** (Inteiro sem sinal de 8 bits). Um número de 8 bits pode armazenar valores de $0$ até $2^{8}-1$, ou seja, **0 a 255**. É por isso que os pixels têm exatamente esse limite de intensidade!

>⚠️ **Atenção:** Se você tiver um pixel com valor 250 e somar 10 a ele usando matemática básica do Numpy, como o limite é 255, ele pode mudar drasticamente devido ao que chamamos de `overflow` (ou seja, ele 'dá a volta' e recomeça a contar a partir de 0, assumindo um valor baixo como 4 em vez de 260). Fique atento a isso quando for clarear imagens no futuro!

## O que podemos fazer com essas matrizes?

### 4. Realce de imagem e filtros
Muitas vezes, a imagem que capturamos de uma câmera ou de um dataset não está com a iluminação ideal. Filtros e técnicas de realce nos ajudam a "limpar" a imagem para que os algoritmos de detecção (ou mesmo nós, humanos) possamos enxergar melhor os detalhes. Vamos explorar três técnicas focadas principalmente no uso sobre matrizes em escala de cinza:

#### 4.1 Equalização de Histogramas
**Para que serve?** Serve para aumentar de forma "bruta" o contraste *global* da imagem. Se uma imagem está muito escura ou muito clara (com os valores de pixel agrupados todos num espaço pequeno da escala de 0 a 255), a equalização abre e distribui essas intensidades preenchendo o espectro inteiro.

**Onde usar?** Quando a imagem tem baixo contraste e você precisa evidenciar detalhes (muito comum para recuperar radiografias escuras, por exemplo).

```python
# A equalização simples atua em imagens de 1 canal (nossa imagem_gray)
imagem_equalizada = cv2.equalizeHist(imagem_gray)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagem_gray, cmap='gray')
plt.title('Original (Baixo Contraste)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagem_equalizada, cmap='gray')
plt.title('Equalização Global')
plt.axis('off')
plt.show()
```

#### 4.2 CLAHE (Contrast Limited Adaptive Histogram Equalization)
**Para que serve?** Diferente da equalização anterior que reajusta a imagem inteira de forma única, o CLAHE é um algoritmo esperto que divide a imagem em pequenos blocos (padrão 8x8) equalizando o contraste *localmente* dentro de cada bloquinho. O "Contrast Limited" impede o contraste extremo, evitando amplificação do ruído.

**Onde usar?** Quando uma imagem tem regiões intensamente iluminadas e, ao mesmo tempo, regiões profundas de sombra. Se usássemos a equalização global, algumas partes "estourariam". O CLAHE recupera o contraste local mantendo a harmonia.

```python
# Cria o objeto CLAHE limitando o contraste (clipLimit) e definindo os blocos
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
imagem_clahe = clahe.apply(imagem_gray)

plt.imshow(imagem_clahe, cmap='gray')
plt.title('CLAHE (Adaptativa)')
plt.axis('off')
plt.show()
```

#### 4.3 Limiarização Adaptativa (Adaptive Threshold)
**Para que serve?** O Threshold transforma a imagem de cinza em uma imagem **Binária** (estritamente Preto e Branco (0 e 255)). Ao invés de usar uma única linha de corte para a imagem toda, ele avalia as proximidades vizinhas de cada pixel para entender se ele deve virar preto ou branco perante aquele contexto local!

**Onde usar?** É extremamente vital em leitura de documentos (OCR), decodificação de códigos de barras ou placas de carro sob sombras fortes ou variações de iluminação, garantindo que o texto/objeto não se perca inteiramente com a claridade desigual.

Três parâmetros importantes no `cv2.adaptiveThreshold`:
* **adaptiveMethod**: Forma do cálculo adaptativo. Pode ser a média da vizinhança (`cv2.ADAPTIVE_THRESH_MEAN_C`) ou soma ponderada gaussiana (`cv2.ADAPTIVE_THRESH_GAUSSIAN_C`).
* **blockSize**: Tamanho da sub-região de pixels usada para calcular (ex: 11x11, sempre ímpar).
* **C**: Valor constante subtraído da média. Reduz um pouco os falsos positivos de limiarização.

```python
# Parâmetros: imagem, max_value, adaptive_method, threshold_type, block_size, C
limiar_adaptativo = cv2.adaptiveThreshold(
    imagem_gray, 
    255, 
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    cv2.THRESH_BINARY, 
    11, 
    2
)

plt.imshow(limiar_adaptativo, cmap='gray')
plt.title('Limiarização Binária (Adaptive Threshold)')
plt.axis('off')
plt.show()
```
